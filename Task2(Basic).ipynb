```python
#predicting the house prices with a different dataset as in the given dataset the column names are missing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
price=pd.read_csv('Book1.csv')
price.head()#loading the dataset and the datset is preprocessed from first only and now we can apply algorithmns on it
```

```python
X=price['Size (sqft)']#feature variable
y=price['Price ($)']#target variable
plt.scatter(X,y)
plt.xlabel('Size')
plt.ylabel('Price of Houses')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
plt.scatter(X_train,y_train,label='Train Data')
plt.scatter(X_test,y_test,label='Test Data')
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr
```

```python
# Clean y_train by removing commas and converting to float
y_train_clean = y_train.replace({',': ''}, regex=True).astype(float)

# Fit the model
lr.fit(X_train.to_numpy().reshape(-1, 1), y_train_clean)
```

```python
y_pred=lr.predict(X_test.to_numpy().reshape(-1,1))
y_pred
```

```python
plt.plot(X_test,y_pred)
```

```python
b1=lr.coef_
b1
```

```python
b0=lr.intercept_
b0
```

```python
print('The equation of the regression line:','y=',b0,'+X',b1)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
# Clean y_test by removing commas and converting to float
y_test_clean = y_test.replace({',': ''}, regex=True).astype(float)

# Evaluate the model using y_test_clean
mae = mean_absolute_error(y_test_clean, y_pred)
mse = mean_squared_error(y_test_clean, y_pred)
r2 = r2_score(y_test_clean, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R Score:", r2)
```

```python
#MAE represents the average absolute difference between the predicted and actual values. A very small MAE close to zero indicates that the model's predictions are extremely accurate. The value of 1.83e-10 is so small that it suggests the model's predictions are almost identical to the actual values.
#MSE is the average of the squared differences between the predicted and actual values. Like MAE, a smaller value indicates better performance. The value of 6.56e-20 is extraordinarily small, which suggests that the model's errors are almost negligible.
#The R score represents the proportion of variance in the dependent variable that is predictable from the independent variables. A perfect R score of 1.0 means that the model is explaining 100% of the variance in the data. This is a strong indicator that the model is perfectly fitting the data.
```
